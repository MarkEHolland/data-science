# Interpreting the FTTP Propensity to Buy Catboost model
---
<span style="background-color:#F5ECCE"> **Specific goal:** For each premise, identify the most likely dominant decision parameter behind the given score</span>, eg: affluence, poor speed, no competition. 
(Although the notebook was created in mind with finding the single most important feature, it can be useful for general explainability.)

This notebook describes a method for identifying key driving factors based on **Shapley values** with the help of the SHAP package: for each premise, finding the most important feature in the local interpretation. </br>

**Other methods** also tried for the same model, but not included in this notebook are: 
- *Local surrogate models* (Lime) - Slow, achieves bad fit locally.
- *Global surrogate models* - Achieves bad fit globally, hard to make interpretations on the level of data points (premises).
- *Model independent methods* - No theoretical support, small difference in technique chosen can lead to very different results.

</br> 

**Shapley values**: A solution concept for calculating the contribution of each player in a coalition game. This answers a question of how to *fairly* distribute payoff among members of a collaborating group. 

Application to feature ML explainability: *How to attribute importance to features that "collaborate" to give us predictions?*
1. The feature values enter a room in random order.
2. The Shapley value of a feature value is the average change in the prediction that the coalition already in the room receives when the feature value joins them.

How to get the value of the prediction when only a subset of the features should *be in the room*?
> For features in the room we take the values from the original observations, and for the features not in the room, we take the values from a random instance (a single random instance per datapoint).

</br> 

**Advantages/Disadvantages**

| Advantages | Disadvantages |
|------------|---------------|
| Strong theoretical support (additivity, local accuracy, consistency, missingness)         |     Computationally intensive (Exact computation is NP-hard)         |
| Approximate methods exist           |   Approximate methods do not have the strong theoretical support            |
| For tree-based models, exact calculation is possible in polinomial time: TreeExplainer (**Lundberg et. al., 2020**)  |     TreeExplainer is still slow          |
| Uses log odds istead of probability by default | You can use model_output="probability" to get SHAP values in probability space (available for xgboost, but not for catboost) |
| Interaction effects are captured (cannot be done by local explanations) | Cannot be used to make statements about changes in prediction for changes in the input |
| Impartial feature credit assignment regardless of tree depth |  Correlation causes problems: Features that have no influence on the prediction function f can get a TreeSHAP estimate different from zero. Also causes inclusion of unrealistic data instances.|
|Consistency with human intuition (claimed in research paper, not generally accepted)|  |

**Characteristics**

Results from game theory imply the Shapley values are the only way to satisfy three important properties (fairness): 
- *Additivity*: when approximating the original model f for a specific input x, the explanation’s attribution values should sum up to the output f(x). Allows for calculating values by trees in a random forest and summing them up.
- *Consistency*: if a model changes so that some feature’s contribution increases or stays the same regardless of the other inputs, that input’s attribution should not decrease. 
- *Missingness*: means that missing or constant feature values should get 0 atribution, it is a trivial property. 


**The interpretation of the Shapley value is**: Given the current set of feature values, the contribution of a feature value to the difference between the actual prediction and the mean prediction is the estimated Shapley value.

Note, you can do much more with SHAP than what is in this notebook, the package is also constantly updated, it is worthwhile to have a look around what is out there for your specific problem.

</br>

```python
def get_key_factor(shap_values, y_pred, cols, row_num):
    """
    Finds the largest shap value in the direction of the prediction:
        - largest positive for positive predicted label
        - largest absolute value amongst the negative shapley values 
        for negative predicted label.
    """
    shap_v_row = shap_values[row_num]
    if y_pred[row_num] == 0:
        most_imp_col_name = cols[np.where(shap_v_row == min(shap_v_row))[0][0]]
    else:
        most_imp_col_name = cols[np.where(shap_v_row == max(shap_v_row))[0][0]]
    return most_imp_col_name
```

```python
def get_stratified_sample(df, target_column, sample_size=-1):
    """
    Returns a sampled dataframe (X) and a series (y). X includes all 
    feature columns with and has a size specified by sample_size. 
    The sampling is stratified by the target. The series y includes 
    the associated values of the target.
    """
    X_all = df.loc[:, df.columns != target_column[0]]
    y_all = df.loc[:, target_column[0]]
        
    if sample_size == -1:
        return X_all, y_all
    else:
        X, _, y, _ = train_test_split(X_all, y_all,
                                stratify=y_all, 
                                test_size=len(X_all) - sample_size,
                                random_state=123)
        return X, y
```

```python
def cb_approximate_shap(X, y, cb_model, categorical_columns):
    """
    Returns the approximate Shapley values for a Catboost model on the given data
    """
    data = Pool(X, y, cat_features=categorical_columns)
    shap_results = cb_model.get_feature_importance(data=data, 
                                                       fstr_type='ShapValues', 
                                                       shap_calc_type="Approximate")
    
    # The last value in each row is the expected value, which is not needed here
    shap_values = shap_results[:,:-1]
    return shap_values
```

```python
def create_key_factors_df(key_factors, key_factors_appr, y_pred):
    """
    Counts how frequently a feature is identified as key driving factor for
    each feature, for both Shapley value calculation methods.
    The inputs are lists containing the key driving factor for each premise 
    with the two methods.
    """
    
    def create_temp_for_label_and_method(data, appr, label=0):
        temp0 = data[data['predicted_value']==label]
        temp = pd.DataFrame(temp0.value_counts(normalize=True), 
                                                 columns=(['temp']))
        temp.reset_index(inplace=True)
        temp.drop(columns=['predicted_value'], inplace=True)
        if appr == False:
            temp.columns = ['Key factor', 
                            'Frequency of being top key factor (regular)']
        else:
            temp.columns = ['Key factor', 
                            'Frequency of being top key factor (approximate)']
        return temp
        
    
    key_factors_df = pd.DataFrame(key_factors, columns=['feature'])
    key_factors_df['predicted_value'] = y_pred
    
    # regular, negative label
    temp1_neg = create_temp_for_label_and_method(key_factors_df, 
                                                 appr=False, 
                                                 label=0)
    # regular, positive label
    temp1_pos = create_temp_for_label_and_method(key_factors_df, 
                                                 appr=False, 
                                                 label=1)
    
    key_factors_aprr_df = pd.DataFrame(key_factors_appr, columns=['feature'])
    key_factors_aprr_df['predicted_value'] = y_pred
    
    # regular, negative label
    temp2_neg = create_temp_for_label_and_method(key_factors_aprr_df, 
                                                 appr=True, 
                                                 label=0)
    # regular, positive label
    temp2_pos = create_temp_for_label_and_method(key_factors_aprr_df, 
                                                 appr=True, 
                                                 label=1)
    
    key_factors_pos_label = temp1_pos.merge(temp2_pos, on='Key factor', how='left').fillna(0)
    key_factors_neg_label = temp1_neg.merge(temp2_neg, on='Key factor', how='left').fillna(0)
    
    return key_factors_pos_label, key_factors_neg_label
```

```python
def plot_key_feature_distribution(key_factors_df, sample_size, num_features, label_type='positive'):
    """
    Plots the ratio of the times when a given feature is identified as the key driving
    factor for the approximate and the regular method for calculating Shapley values.
    The ratios are stored in key_factors_df.
    The number of features shown on the plot is controlled by num_features. The features 
    are shown in order of how often they are identified as key driving factor by the regular
    method.
    """
    key_factors_df[:num_features].plot(x='Key factor', 
                                       y=['Frequency of being top key factor (regular)',
                                          'Frequency of being top key factor (approximate)'],
                                       kind="bar", 
                                       figsize=(15,4))
    ax = plt.gca()
    ax.set_ylabel('Frequency', fontsize=15)
    ax.xaxis.label.set_size(15)
    ax.tick_params(axis='both', which='major', labelsize=15)
    temp = [round(x/10,1) for x in range(1,11)]
    ticks = [x for x in temp if x < max(key_factors_df.iloc[:,1].max(),
                                                                  key_factors_df.iloc[:,2].max())]
    ax.set_yticks(ticks)
    ax.set_yticklabels(["{0:.0%}".format(t) for t in ticks])
    ax.legend(fontsize=15)
    plt.title(f"Distribution of the key factor identified by the regular and approximate method"
              f"\n for {label_type} labels"
              f"\n (top {num_features} features included, sample size: {sample_size})",
             fontsize=15)
    plt.show()
    return
```

```python
def create_dependence_plot(shap_values, X, col, ax, interaction_index, y_min=-6, y_max=6):
    ax.set_ylim(y_min,y_max)
    # horizontal line at zero, for a visual aid
    ax.axhline(y=0.0, color='k', linestyle='-', linewidth=0.5)
    if col in categorical_columns or col.endswith("_flag") or col.endswith("_ordinal"):
        # x jitter for better visibility
        x_jitter = 0.4
    else:
        x_jitter = 0
    shap.dependence_plot(col, shap_values, X, interaction_index=interaction_index, show=False, x_jitter=x_jitter, ax=ax)
    return

def dependence_plot_matrix(shap_values, X, cols=None, interaction_index='auto', in_same_line=4, title=True, more_space=False):
    if cols == None:
        cols = X.columns
    num_of_lines = math.ceil(len(cols)/in_same_line)
    if more_space:
        fig_height = 12
    else:
        fig_height = 6
    fig, axs = plt.subplots(num_of_lines, in_same_line, figsize=(15,num_of_lines*5))
    ax_iter = iter(axs.flat)
    for col in cols:
        ax = next(ax_iter)
        create_dependence_plot(shap_values, X, col, ax, interaction_index)
    num_empty_axes = num_of_lines * in_same_line - len(cols)
    for i in range(num_empty_axes):
        axs.flat[(-1) * (i+1)].set_visible(False) 
    plt.tight_layout()
    
    if interaction_index == None:
        interaction_desc = "without interactions"
    if interaction_index == 'auto':
        interaction_desc = "without automatic best interactions"
    if interaction_index in list(X.columns):
        interaction_desc = f"with interactions with the feature: {interaction_index}"
    if title:
        plt.suptitle(f"Dependence plot matrix {interaction_desc}\n (sample size: {sample_size})", y=1.02, fontsize=18)
    plt.show()
    
```

## Validate approximate method of collecting Shapley values
The best way for calculating Shap values for a tree-based model, is the `shap.TreeExplainer`. This method, however, is very slow, and therefore not practical for large datasets.
There is an alternative approximate method which does not have the consistency guarantees of the regular method. Here, we compare the two methods in terms of the identified most important feature for each data point.

```python
# sample size for comparing the approximate and the regular methods. For the sample size of 100, the regular method takes 2 minutes.
sample_size = 500
print(f"The sample size for the comparison is: {sample_size}.")

# Get a sample
X, y = get_stratified_sample(training_df_tra, target_column, sample_size)

# Get predictions
y_pred = cb_model.predict(X)
y_probs = cb_model.predict_proba(X)
```

```python
%%time

# Calculating shap values for the sample by a regular TreeExplainer.
explainer = shap.TreeExplainer(cb_model)
shap_values = explainer.shap_values(X)
print(f"Time to calculate Shapley values with the \033[1mregular TreeExplainer\033[0m for a a sample with a size of {sample_size}:")
```

```python
%%time

#Calculating shap values for the sample by the approximate method implemented in the Catboost package.
shap_values_appr = cb_approximate_shap(X, y, cb_model, categorical_columns)
print(f"Time to calculate \033[1mapproximate\033[0m Shapley values for a a sample with a size of {sample_size}:")
```

```python
# Get key factor for each data point 
key_factors = [get_key_factor(shap_values, y_pred, list(X.columns), i) 
                     for i,e in enumerate(X.index)]
key_factors_appr = [get_key_factor(shap_values_appr, y_pred, list(X.columns), i) 
                     for i,e in enumerate(X.index)]
```

```python
differences = ['Different' if a!=b else 'Same' for a,b in zip(key_factors, key_factors_appr)]
agreement_prcnt = differences.count('Same')/len(differences)*100
print(f"\033[0;30;43m The regular and the approximate method agrees {agreement_prcnt:.2f}% "
      f"of the time in a sample with a size of {sample_size}.")
```

### Dependence plots

Plots the value of the feature on the x-axis and the SHAP value of the same feature on the y-axis. This shows how the model depends on the given feature, and is like a richer extenstion of the classical partial dependence plots. 
It is possible to also show interaction of the given feature with one another feature in the plot. 

### Dependence plots with original feature values without interaction
Each of the plots is a scatterplot of a feature and the associated Shapley values. Shapley values above 0 mean that the feature value in the given the other features describing the datapoint is associated with a positive effect, that is, a higher chance of conversion. A Shapley value below 0 is associated with a lower chance of conversion.
The same value of a feature might be associated with different Shapley values for different datapoints, depending on the rest of the features.

## Grouping features with related meaning

```python
groups_of_features = {
    'product_availability': ['notspot_to_fttp_less_30mbps_flag', 
                       'gfast_avl_flag', 
                       'adsl_avl_flag', 
                       'fttc_avl_flag',],
    'faults': ['had_fault_ever_flag', 
               'had_fault_in_last_six_months_flag', 
               'had_fault_in_last_year_flag', 
               'n_exch_faults_in_last_year'],
    'premise_characteristics': ['site_classification_desc_binned', 
                                'premise_class',
                                'build_category'],
    'assets_at_ord': ['ever_had_asset_at_ord_flag', 
                      'has_asset_at_ord_flag', 
                      'act_FTTC_at_ord_flag', 
                      'act_MPF_at_ord_flag',
                      'act_SMPF_at_ord_flag', 
                      'act_SOGEA_at_ord_flag',],
    'last_asset': ['last_sub_product_binned', 
                   'last_technology', 
                   'months_since_no_asset_capped'],
}

features_one_by_one = ['fttp_network_type_CBT_flag', 'best_fibre_dp_type',
                   'is_greenfield_flag', 'ofcom_area2_flag',
                   'geo_type_ordinal', 'quarters_since_emlc_ok',
                   'quarters_since_last_order',  'age_ordinal',
                   'household_income_ordinal', 'digital_ordinal', 
                   'has_made_eoi_flag', 'is_competitor_at_postcode_flag']

for col in features_one_by_one:
    groups_of_features[col] = [col]
```

```python
revert_dict = lambda d: dict(chain(*[zip(val, repeat(key)) for key, val in d.items()]))

def grouped_shap(shap_vals, features, groups):
    groupmap = revert_dict(groups)
    shap_Tdf = pd.DataFrame(shap_vals, columns=pd.Index(features, name='features')).T
    shap_Tdf['group'] = shap_Tdf.reset_index().features.map(groupmap).values
    shap_grouped = shap_Tdf.groupby('group').sum().T
    return shap_grouped
```
