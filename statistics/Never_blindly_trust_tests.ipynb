```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kstest, norm, cauchy
```

```python
# define sample space (quantiles)
sample_space = np.arange(-30, 30, 0.01)
# plot the probability density function for different degrees of freedom values
fig = plt.figure(figsize=(8,5))
plt.plot(sample_space, norm(loc=10, scale=1).pdf(sample_space), label='Gaussian')
plt.plot(sample_space, cauchy(loc=-10, scale=.5).pdf(sample_space), label='Cauchy')
plt.title('Probability density functions')
plt.legend()
plt.show()
```

```python
kstest(rvs=cauchy(loc=-10, scale=.5).rvs(size=20000, random_state=123), cdf=norm(loc=10, scale=1).cdf)
```

```python
kstest(rvs=cauchy(loc=-10, scale=.5).rvs(size=20000, random_state=123), cdf=norm(loc=10, scale=1).cdf, alternative='less')
```
