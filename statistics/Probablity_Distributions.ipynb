```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kstest, norm, cauchy
```

```python
cau = cauchy(loc=10, scale=1)
no = norm(loc=10, scale=1)
```

```python
n_samples = 100000
```

```python
means_cau = []
stds_cau = []

means_no = []
stds_no = []
```

```python
samples_cau = cau.rvs(size=n_samples, random_state=123)
samples_no = no.rvs(size=n_samples, random_state=123)
```

```python
for i in range(n_samples):
    means_cau.append(np.mean(samples_cau[:i]))
    stds_cau.append(np.std(samples_cau[:i]))
    
    means_no.append(np.mean(samples_no[:i]))
    stds_no.append(np.std(samples_no[:i]))
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(16,5))
# Means
ax1.plot(np.arange(n_samples), means_cau, label='Cauchy')
ax1.plot(np.arange(n_samples), means_no, label='Gaussian', color='orange')
ax1.set_title('Means')
ax1.set_xlabel('N. Samples')
ax1.legend()
# St.Dev.
ax2.plot(np.arange(n_samples), stds_cau, label='Cauchy')
ax2.plot(np.arange(n_samples), stds_no, label='Gaussian', color='orange')
ax2.set_title('St.Dev.')
ax2.set_xlabel('N. Samples')
ax2.legend()

plt.show()
```
